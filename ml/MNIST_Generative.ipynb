```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(y_train, x_train), (y_test, x_test) = keras.datasets.mnist.load_data()
y_train = y_train.reshape(y_train.shape[0], 784).astype('float32') / 255.0
y_test = y_test.reshape(y_test.shape[0], 784).astype('float32') / 255.0

x_train = keras.utils.to_categorical(x_train, 10)
x_test = keras.utils.to_categorical(x_test, 10)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_train[0])
print(y_train[0])
```

```python
img = y_train[0].reshape((28,28))
plt.imshow(img,cmap='grey')
```

```python
model = keras.Sequential([
        keras.layers.Dense(256, activation='relu', input_shape=(10,)),
        keras.layers.Dense(512, activation='relu'),
        keras.layers.Dense(784, activation='sigmoid')  # Output: flattened image (28x28)
    ])

model.compile(optimizer='adam', loss='mse')  # Mean Squared Error loss
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=128)
```

```python
def generate_image(number):
    input_number = np.zeros((1, 10))
    input_number[0, number] = 1
    generated_image = model.predict(input_number)
    generated_image = generated_image.reshape(28, 28)
    return generated_image
```

```python
import random


for i in range(20):
  n = random.randint(0, 9)
  generated_n = generate_image(n)
  print("Generated: ", n)
  plt.imshow(generated_n, cmap='gray')
  plt.show()
```
